> **[12]** `horsepower` 변수를 z-score 표준화를 진행하고 이상치 수 구하기.

- 이상치는 `z`값이 `1.5`를 초과하거나 `-1.5` 미만인 값
$$z = \frac{(x - 평균)}{표준편차} $$ 

````python
df1 = df.copy()

# z-score 표준화 하기
meanValue = df1['horsepower'].mean()
stdValue = df1['horsepower'].std()

print(meanValue, stdValue)

df1['hp_horse_zscore'] = (df1['horsepower'] - meanValue) / stdValue
print(df1.head(2))

# 이상치 구하기
# (참고) 2개 이상의 연산을 합칠 때는, 연산 마다 괄호로 묶어준다.
cond1 = df1['hp_horse_zscore'] > 1.5
cond2 = df1['hp_horse_zscore'] < -1.5

answer = len(df1[cond1 | cond2])
print(f"answer: {answer}")
````

> **[13]** `mpg` 컬럼을 Min-Max 정규화를 진행한 후, 0.8보다 큰 값을 가지는 레코드 수 구하기

$$MinMax = \frac{x - Min}{Max - Min}$$

````python
df1 = df.copy()

# Min-Max 정규화 진행하기
minValue = df1['mpg'].min()
maxValue = df1['mpg'].max()

df1['mpg_minmax'] = (df['mpg'] - minValue) / (maxValue - minValue)
print(df1.head(3))

# 0.8보다 큰 값을 가지는 레코드 수 구하기
cond = df1['mpg_minmax'] > 0.8

answer = len(df1[cond])
print(f"answer: {answer}")
````

> **[14]** `weight` 컬럼에 대해 상자그림(박스플롯) 기준으로 이상치 개수 구하기

```text
IQR = Q3 - Q1

UPPER_BOUND = Q3 + 1.5 * IQR
LOWER_BOUND = Q1 - 1.5 * IQR
```

- 이상치는 다음의 범위에 해당하는 값에 해당한다.
    - `x > UPPER_BOUND`
    - `x < LOWER_BOUND`

````python
df1 = df.copy()

# IQR, Upper Bound, Lower Bound 구하기
Q1, Q3 = df1['weight'].quantile([0.25, 0.75])
IQR = Q3 - Q1
print(Q1, Q3, IQR)

UPPER_BOUND = Q3 + (1.5 * IQR)
LOWER_BOUND = Q1 - (1.5 * IQR)
print(UPPER_BOUND, LOWER_BOUND)

# 이상치 개수 구하기
cond1 = df1['weight'] > UPPER_BOUND
cond2 = df1['weight'] < LOWER_BOUND

answer = len(df1[cond1 | cond2])
print(f"answer: {answer}")
````

> **[참고]** `mpg` 데이터의 이상치를 평균값으로 대체하기

````python
data = {
    'mpg': [18, 22, 30, 25, 50, 24, 19, 150, 26, 21],
    'horsepower': [130, 165, 150, 140, 198, 150, 180, 200, 160, 170]
}
ref_df = pd.DataFrame(data)

# IQR, LB, UP 구하기
Q1, Q3 = ref_df['mpg'].quantile([0.25, 0.75])
IQR = Q3 - Q1
LOWER_BOUND = Q1 - 1.5 * IQR
UPPER_BOUND = Q3 + 1.5 * IQR

# 이상치 확인하기
cond1 = ref_df['mpg'] < LOWER_BOUND
cond2 = ref_df['mpg'] > UPPER_BOUND

outliers = ref_df['mpg'][cond1 | cond2]
print(outliers)  # 4번째, 7번째 인덱스 값이 이상치

print()

# 이상치 대체하기 (평균값)
meanValue = outliers.mean()
print(meanValue)

print()

## 1. NumPy를 이용하는 방법
import numpy as np
ref_df['mpg'] = np.where((cond1 | cond2), meanValue, ref_df['mpg'])
print(ref_df)

print()

## 2. Pandas를 이용하는 방법
ref_df.loc[(cond1) | (cond2), 'mpg'] = meanValue
print(ref_df)
````

````python
df1 = df.copy()

# 매출액 구하기
df1['매출액'] = df1['판매량'] * df1['개당수익']
# print(df1)

print()

# 날짜형으로 변환하기 : pd.to_datetime()
df1['날짜'] = pd.to_datetime(df1['날짜'])
print(df1.info())

print()

## 방법 1 : .dt.[year | quarter | month | day]
df1['년'] = df1['날짜'].dt.year
df1['분기'] = df1['날짜'].dt.quarter
df1['월'] = df1['날짜'].dt.month
df1['일'] = df1['날짜'].dt.day

print(df1)

print()

cond1 = df1['년'] == 2022
cond2 = df1['분기'] == 1
cond3 = df1['제품'] == '사과'

answer = df1['매출액'][cond1 & cond2 & cond3].sum()
print(f"answer: {answer}")

## 방법 2 : 날짜 컬럼 직접 사용하기
cond1 = df1['날짜'].between('2022-01-01', '2022-03-31')
cond2 = df1['제품'] == '사과'

answer = df1['매출액'][cond1 & cond2].sum()
print(f"answer: {answer}")
````

> **[26]** 22년과 23년 총 매출액의 차이를 절댓값으로 구하기

> **[27]** 23년 총 매출액이 큰 제품의 23년 판매량 구하기

````python
df1 = df.copy()

# 매출액 컬럼 만들기
df1['매출액'] = df1['판매량'] * df1['개당수익']

# 판다스 날짜 형식으로 변환하기
df1['날짜'] = pd.to_datetime(df1['날짜'])

# 년 컬럼 만들기
df1['년'] = df1['날짜'].dt.year

# 매출액이 가장 높은 제품 구하기
a = df1.groupby(by='제품')['매출액'].sum()
print(a)

maxProduct = a.idxmax()  # 값이 가장 큰 레코드의 인덱스값 가져오기
print(maxProduct)

print()

# 매출액이 가장 큰 제품의 23년 판매량 구하기
cond1 = df1['년'] == 2023
cond2 = df1['제품'] == maxProduct

answer = df1['판매량'][cond1 & cond2].sum()
print(f"answer: {answer}")
````
