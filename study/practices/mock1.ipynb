### <a id='toc1_1_2_'></a>[문제 2 (10점)](#toc0_)
---

- `telco-customer-churn` 데이터 세트(`M1-2.csv`)
- `TotalChanges` 항목에서 결측값을 제거하고 이상값을 제외한 평균을 정수로 출력하기
    - 이상값은 평균에서 1.5 표준편차 이상인 값으로 하기

```python
import pandas as pd
import numpy as np

d_cust = pd.read_csv('./datasets/M1-2.csv')

# 'TotalCharges' 열의 데이터를 숫자로 변환
d_cust['TotalCharges'] = pd.to_numeric(d_cust['TotalCharges'], errors='coerce')

v_mean = np.nanmean(d_cust['TotalCharges'])    # 결측값을 제거하고 평균값 구하기
v_sd = np.nanstd(d_cust['TotalCharges'])   # 결측값을 제거하고 표준편차 구하기
v_upper = v_mean + v_sd * 1.5   # 1.5 표준편차
v_under = v_mean - v_sd * 1.5   # 1.5 표준편차

# 평균에서 1.5 표준편차 이상인 값으로 하기
crit = (d_cust['TotalCharges'] >= v_under) & (d_cust['TotalCharges'] <= v_upper)
filtered_data = d_cust[crit]

# 평균 구하기
summary = filtered_data['TotalCharges'].mean()   

# 정수로 출력하기
print(int(summary))
```

### <a id='toc1_1_3_'></a>[문제 3 (10점)](#toc0_)
---
- 32개 자동차 모델의 디자인과 성능을 비교한 `mtcars` 데이터 (`M1-3.csv`)
- 수동(`am=1`) 중에서 가장 마력(`hp`)이 작은 5개의 데이터의 평균 연비(`mpg`)와 자동(`am=0`) 중에서 가장 마력(`hp`)이 작은 5개 데이터의 평균 연비(`mpg`)의 차이 구하기
- 소수점 둘째 자리에서 반올림하여 소수점 첫째 자리로 출력

```python
import pandas as pd

ds = pd.read_csv('./datasets/M1-3.csv')

# (1) 수동(am=1)
## 마력(hp) 기준으로 오름차순 정렬 후, 상위 5개 값 추출
crit1 = (ds['am'] == 1)   # 수동
ds_man_min = ds[crit1].sort_values(by='hp', ascending=True).head(5)   
print(ds_man_min)

print()

## 연비(mpg)의 평균 구하기
mean_man_mpg = ds_man_min['mpg'].mean()   
print(mean_man_mpg)

print()

# (2) 자동(am=0)
## 마력(hp) 기준으로 오름차순 정렬 후, 상위 5개 값 추출
crit2 = (ds['am'] == 0)
ds_aut_min = ds[crit2].sort_values(by='hp', ascending=True).head(5)   
print(ds_aut_min)

print()

## 연비(mpg)의 평균 구하기
mean_aut_mpg = ds_aut_min['mpg'].mean()
print(mean_aut_mpg)

print()

# (3) 차이 구하기 (수동 - 자동)
result = round(mean_man_mpg - mean_aut_mpg, 1)   # 소수점 첫째 자리로 출력
print(result)
```

### <a id='toc1_2_1_'></a>[문제 (40점)](#toc0_)
---
- `telco-customer-churn` 데이터
    - 각 고객이 가입한 서비스와 계정 정보, 인구에 대한 통계 정보들
- 주어진 훈련 데이터(`M1-4-1.csv`)를 이용하여 모델을 훈련한 후 테스트 데이터(`M1-4-2.csv`)로 고객의 이탈 여부를 예측하고 `csv` 포맷으로 만들기
- 단, 이탈: `Yes`, 유지: `No`

```python
import pandas as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score

# (1) 훈련용 데이터 불러오기 & 확인
df1 = pd.read_csv('./datasets/M1-4-1.csv')

## Churn 컬럼 값 수정하기
df1['Churn'] = df1['Churn'].map({ 'No': 0, 'Yes': 1 })   # 문자형을 수치형으로 바꾸기 (결과가 0 또는 1로 나타나도록)

## 요약 통계량 확인
print(df1.describe())

print()
print()

## 속성 및 결측값 확인
print(df1.info())
```

```python
# (2) 평가용 데이터 불러오기 & 확인
df2 = pd.read_csv('./datasets/M1-4-2.csv')

## Churn 컬럼 값 수정하기
df2['Churn'] = df1['Churn'].map({ 'No': 0, 'Yes': 1 })  # 문자형을 수치형으로 바꾸기 (결과가 0 또는 1로 나타나록)

print(df2.describe())

print()
print()

print(df2.info())
```

```python
# (3) 훈련용 데이터 작업
## 데이터 형변환
df1['Churn'] = df1['Churn'].astype('category')   # 수치형 -> 범주형

## 독립 변수, 종속 변수 분리
x = df1.drop('Churn', axis=1)
y = df1['Churn']

## 원-핫 인코딩 (독립 변수)
x_encoded = pd.get_dummies(x)

## 데이터 분할
x_train, x_valid, y_train, y_valid = train_test_split(x_encoded, y, test_size=0.25)

## 모델링
md = RandomForestClassifier(n_estimators=300)
md.fit(x_train, y_train)

## 예측
### 모델을 사용하여 검증 데이터 예측
x_valid_encoded = pd.get_dummies(x_valid)
x_valid_encoded = x_valid_encoded.reindex(columns=x_encoded.columns, fill_value=0)   # x_encoded의 열 순서에 맞게 재인덱싱

pred_valid = md.predict(x_valid_encoded)

## 평가
### 혼동 행렬, 정확도, 재현율, 정밀도, F1 점수, ROC AUC 점수 출력
cm = confusion_matrix(y_valid, pred_valid)
print(cm)

print()

print(accuracy_score(y_valid, pred_valid))   # 정확도
print(recall_score(y_valid, pred_valid))    # 재현율
print(precision_score(y_valid, pred_valid))   # 정밀도
print(f1_score(y_valid, pred_valid))        # F1 점수
print(roc_auc_score(y_valid, pred_valid))   # ROC AUC

# (4) 평가용 데이터 작업
x_test = df2

## 원-핫 인코딩
x_test_eoncoded = pd.get_dummies(x_test)
x_test_eoncoded = x_test_eoncoded.reindex(columns=x_encoded.columns, fill_value=0)   # x_encoded의 열 순서에 맞게 재인덱싱

## 예측
pred_new = md.predict(x_test_eoncoded)
pred_new_label = ['No' if pred == 0 else 'Yes' for pred in pred_new]
print(pred_new_label)

print()

# (5) 결과 출력 및 CSV 내보내기
result = pd.DataFrame({ 'pred': pred_new_label })
print(result)

result.to_csv('./outputs/mock1-section2.csv', index=False)
```

### <a id='toc1_3_1_'></a>[문제 1 (15점)](#toc0_)
---

- 5명의 환자를 대상으로 치료제를 복용하기 전과 후의 혈압을 측정하였다.
- 치료제가 효과가 있는지 쌍체 표본 T-검정을 통해 답하고자 한다.
    - 단, 표본이 정규성을 만족한다는 가정하에 단측 검정 수행

<br/>

$$\mu_{d} : \text{(치료 후 혈압 - 치료 전 혈압)의 평균} \\ H_{0} : \mu_{d} ≥ 0 \text{(치료 후 혈업과 치료 전 혈압의 차이에 대한 평균은 0보다 크거나 같다.)} \\ H_{1} : \mu_{d} < 0 \text{(치료 후에 혈압이 더 낮아졌으므로 치료제의 효과가 있다.)}$$

<br/>

|치료제 복용 전 5명의 환자들의 혈압|200, 210, 190, 180, 175|
|:-:|:-:|
|치료제 복용 후 5명의 환자들의 혈압|180, 175, 160, 150, 160|

<br/>

> 문제

- (1) 위의 가설을 검정하기 위한 검정 통계량 구하기 (반올림하여 소수 다섯째 짜리까지 계산) (5점)
- (2) 위의 통계량에 대한 p-값 구하기 (반올림하여 소수 다섯째 짜리까지 계산) (5점)
- (3) 유의수준 0.05 하에서 가설 검정의 결과를 (채택/기각) 중 하나를 선택하여 나타내기 (5점)

```python
import pandas as pd
from scipy.stats import ttest_rel

# 주어진 값들을 바탕으로 데이터프레임 만들기
df = pd.DataFrame({
    'before': [200, 210, 190, 180, 175],
    'after': [180, 175, 160, 150, 160]
})

print(df)

print()

# 쌍체 표본 T-검정 수행
result_t_test = ttest_rel(df['after'], df['before'], alternative='less')   
# alternative 값
## - two_sided: 양측 검정
## - less: 단측 검정 (표본 평균이 특정 값보다 작은지)
## - greater: 단측 검정(표본 평균이 특정 값보다 큰지)

print(result_t_test)

print()

# 정답 (1) : 검정 통계량
result1 = round(result_t_test.statistic, 5)   # 소수점 다섯째 자리까지 계산
result1 = float(result1)
print(result1)

print()

# 정답 (2) : p-값
result2 = round(result_t_test.pvalue, 5)   # 소수점 다섯째 자리까지 계산
result2 = float(result2)
print(result2)

print()

# 정답 (3) : 유의수준 0.05 하에서 가설검정의 결과를 채택/기각 중 하나로 표시
result3 = "기각" if result2 < 0.05 else "채택"
print(result3)
```

### <a id='toc1_3_2_'></a>[문제 2 (15점)](#toc0_)
---

- 뉴욕의 공기 오염도를 측정한 `airqty` 데이터 세트(`M1-6.csv`)
    - `Temp` 데이터는 온도를 나타낸다.
- `airqty` 데이터 세트에서 다음 가설에 대한 통계적 검정 수행하기 (양측 검정 수행)

|귀무가설($H_{0}$)|뉴욕의 평균 온도(Temp 변수 값의 평균)는 75이다.|
|:-:|-|
|대립가설($H_{1}$)|뉴욕의 평균 온도(Temp 변수 값의 평균)는 75가 아니다.|

<br/>

> 데이터

|변수|설명|
|:-:|-|
|Ozone|오존농도|
|Solar.R|자외선|
|Wind|바람|
|Temp|온도|
|Month|월|
|Day|일|

<br />

> 문제

- (1) 위의 가설을 검정하기 위한 검정 통계량 구하기 (반올림하여 소수 셋째 자리까지 계산) (5점)
- (2) 위의 통계량에 대한 p-값 구하기 (반올림하여 소수 넷째 자리까지 계산) (5점)
- (3) 유의수준 0.05 하에서 가설 검정의 결과를 채택/기각 중 하나를 선택하여 나타내기 (5점)

```python
import numpy as np
import pandas as pd
from scipy.stats import shapiro, wilcoxon

airqty = pd.read_csv('./datasets/M1-6.csv')

# 요약 통계량
print(airqty.describe())

print()
print()

# 샤피로 윌크 검정 수행 (Temp)
result_shapiro = shapiro(airqty['Temp'])
print(result_shapiro)

print()

## p-값이 0.009319356335949125 이므로, 0.05보다 작으므로 정규성 가정을 불만족한다.
## 따라서 윌콕슨 함수를 이용하여 검정을 시행한다.

# 윌콕슨 함수를 이용하여 양측 검정 시행
result_wilcoxon = wilcoxon(airqty['Temp'], alternative='two-sided', zero_method='wilcox', correction=False)
print(result_wilcoxon)

print()

# 정답 (1) : 검정 통계량 구하기
result1 = round(result_wilcoxon.statistic, 3)   # 소수점 셋째 자리까지 계산
result1 = result1.item()
print(result1)

print()

# 정답 (2) : p-값 구하기
result2 = round(result_wilcoxon.pvalue, 4)   # 소수점 넷째 자리까지 계산
print(result2)

print()

# 정답 (3) : 유의수준 0.05 하에서 가설검정의 결과를 채택/기각 중 하나로 표시
result3 = "기각" if result2 < 0.05 else "채택"
print(result3)
```
