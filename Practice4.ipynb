    - `where` í•¨ìˆ˜ë¥¼ ì´ìš©í•´ì„œ íŒŒìƒ ë³€ìˆ˜ `capital_gain_yn`, `capital_loss_yn`ì„ ë§Œë“ ë‹¤.

```python
X_train['capital_gain_yn'] = np.where(X_train['capital_gain'] > 0, 1, 0)
X_train['capital_loss_yn'] = np.where(X_train['capital_loss'] > 0, 1, 0)

X_test['capital_gain_yn'] = np.where(X_test['capital_gain'] > 0, 1, 0)
X_test['capital_loss_yn'] = np.where(X_test['capital_loss'] > 0, 1, 0)
```

- ì¢…ì† ë³€ìˆ˜ ê°’ë³„ë¡œ ë…ë¦½ ë³€ìˆ˜ ê°’ì˜ ì°¨ì´ê°€ ìˆëŠ”ì§€ íƒìƒ‰í•˜ê¸° ìœ„í•´ ë³€ìˆ˜ ì¢…ë¥˜ë³„ë¡œ ë‚˜ëˆ ë³¸ë‹¤.

```python
COL_DEL = []
COL_NUM = ['age', 'education_num', 'hours_per_week', 'capital_gain', 'capital_loss']    # ìˆ˜ì¹˜í˜• ë³€ìˆ˜
COL_CAT = ['workclass', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'native_country', 'capital_gain_yn', 'capital_loss_yn']    # ë²”ì£¼í˜• ë³€ìˆ˜
COL_Y = ['target']

X_train = X_train.drop(COL_DEL, axis=1)
X_test = X_test.drop(COL_DEL, axis=1)
```

```python
train_df = pd.concat([X_train, y_train], axis=1)

for _col in COL_NUM:
    print("-" * 80)
    print(_col)
    print(train_df.groupby(COL_Y)[_col].describe(), end="\n\n")
```

    native country ë³€ìˆ˜ì˜ ê²½ìš° 41ê°œì˜ ê³ ìœ í•œ ê°’ì„ ê°–ëŠ”ë‹¤. (í•œ ë³€ìˆ˜ê°€ ê°€ì§ˆ ìˆ˜ ìˆëŠ” ê³ ìœ í•œ ê°’ì˜ ê°œìˆ˜ë¥¼ <b>ì¹´ë””ë„ë¦¬í‹°(Cardinality)</b>ë¼ê³  í•œë‹¤.) <br/>

    ì¹´ë””ë„ë¦¬í‹°ê°€ ë†’ì„ ê²½ìš°, í•´ë‹¹ ë³€ìˆ˜ë¥¼ ì œê±°í•˜ê±°ë‚˜ ì¼ë¶€ ë²”ì£¼ë¥¼ í•©ì³ ì¹´ë””ë„ë¦¬í‹°ë¥¼ ë‚®ì¶œ ìˆ˜ ìˆë‹¤. ë˜ëŠ” (ë‹¤ë£¨ê¸° ê¹Œë‹¤ë¡­ì§€ë§Œ) íƒ€ê²Ÿ ì¸ì½”ë”©ì„ ì ìš©í•  ìˆ˜ ìˆë‹¤.

- ë²”ì£¼í˜• ë³€ìˆ˜ë“¤ì— ëŒ€í•´ **ë ˆì´ë¸” ì¸ì½”ë”©**ì„ ìˆ˜í–‰í•œë‹¤.
    - ë ˆì´ë¸” ì¸ì½”ë”©ì€ **ë¬¸ì**ë¡œ ëœ ë²”ì£¼ë¥¼ **ì •ìˆ˜** í˜•íƒœë¡œ ë°”ê¿”ì¤€ë‹¤.

```python
from sklearn.preprocessing import LabelEncoder

X = pd.concat([X_train, X_test])
for _col in COL_CAT:
    le = LabelEncoder()
    le.fit(X_train[_col])
    X_train[_col] = le.transform(X_train[_col])
    X_test[_col] = le.transform(X_test[_col])
```

<div class="alert alert-warning">
    ë ˆì´ë¸” ì¸ì½”ë”©ì„ í•œ ë°ì´í„°ëŠ” <u>ì„ í˜• ëª¨ë¸</u>ì—ëŠ” ì í•©í•˜ì§€ ì•Šë‹¤. (ë ˆì´ë¸” ì¸ì½”ë”© í•œ ë³€ìˆ˜ ê°’ì´ ì˜ë¯¸ë¥¼ ê°€ì§€ëŠ” ê²ƒì´ ì•„ë‹ˆê¸° ë•Œë¬¸ì´ë‹¤.) <br/>
    ë”°ë¼ì„œ ë°ì´í„° ëª¨í˜• êµ¬ì¶• ì‹œ, <u>ë¹„ì„ í˜• ëª¨ë¸</u>ì¸ <b>ëœë¤ í¬ë ˆìŠ¤íŠ¸</b>ì™€ <b>XGBoost</b> ì•Œê³ ë¦¬ì¦˜ì„ ì‚¬ìš©í•˜ë„ë¡ í•œë‹¤.
</div>

### ë°ì´í„° ëª¨í˜• êµ¬ì¶•

- ë¶„ë¥˜ ì˜ˆì¸¡ ëª¨í˜•ì„ ë§Œë“¤ê³ ì ìš°ì„  ë°ì´í„° ë¶„í• ì„ ì§„í–‰í•œë‹¤.

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.3, stratify=y_train)
```

- ë¶„í• ëœ ë°ì´í„°ë¡œ **ìˆ˜ì¹˜í˜• ë³€ìˆ˜**ì˜ ë°ì´í„° ìŠ¤ì¼€ì¼ë§ì„ ìˆ˜í–‰í•œë‹¤.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_tr[COL_NUM] = scaler.fit_transform(X_tr[COL_NUM])
X_val[COL_NUM] = scaler.fit_transform(X_val[COL_NUM])
X_test[COL_NUM] = scaler.fit_transform(X_test[COL_NUM])
```

- splitì´ ë˜ì—ˆìœ¼ë‹ˆ ë°”ë¡œ í•´ë‹¹ ë°ì´í„°ë¥¼ ì‚¬ìš©í•˜ì—¬ í•™ìŠµì„ ì§„í–‰í•œë‹¤.
- `sklearn` íŒ¨í‚¤ì§€ë¥¼ ì´ìš©í•˜ì—¬ ëœë¤ í¬ë ˆìŠ¤íŠ¸, XGBoost ëª¨ë¸ì„ ë§Œë“¤ì–´ë³´ì.

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(X_tr, y_tr.values.ravel())
```

```python
from xgboost import XGBClassifier

model_xgb1 = XGBClassifier()
model_xgb1.fit(X_tr, y_tr.values.ravel())
```

<div class="alert alert-warning">
    XGBoost í•™ìŠµ ì‹œ, <code>eval_set</code>ì„ ì‚¬ìš©í•˜ì—¬ ì¶©ë¶„íˆ í•™ìŠµì„ ì‹œí‚¤ë©´ì„œ ê³¼ì í•©ì„ ë°©ì§€í•˜ë„ë¡ í•  ìˆ˜ ìˆë‹¤.
</div>

```python
model_xgb2 = XGBClassifier(n_estimators=1000, learning_rate=0.1, max_depth=10)
model_xgb2.fit(X_tr, y_tr.values.ravel(), early_stopping_rounds=50, eval_metric='auc', eval_set=[(X_val, y_val)], verbose=10)
```

<div class="alert alert-warning">
    n_estimators ê°’ì„ 1,000ìœ¼ë¡œ í¬ê²Œ ì¡ì„ ê²½ìš° í•™ìŠµì„ ë” ì˜ í•  ìˆ˜ ìˆì§€ë§Œ, ê·¸ë§Œí¼ ê³¼ëŒ€ì í•© ìœ„í—˜ì´ í¬ë‹¤. <br/>
    ê·¸ë˜ì„œ fitting í•  ë•Œ eval_setê³¼ early_stopping_rounds ê°’ì„ ì§€ì •í•´ì£¼ì—ˆë‹¤. <br/>
    ëª¨ë¸ì´ ë§¤ iteration ë§ˆë‹¤ eval_setì„ ì´ìš©í•˜ì—¬ ì„±ëŠ¥ì„ ê³„ì‚°í•˜ë‹¤ê°€ early_stopping_rounds ì´ìƒì˜ iteration ë™ì•ˆ ì„±ëŠ¥ ê°œì„ ì´ ì—†ìœ¼ë©´ í•™ìŠµì„ ë©ˆì¶˜ë‹¤.
</div>

- ì´ˆë„ ëª¨ë¸ì˜ ì„±ëŠ¥ì„ í™•ì¸í•˜ê¸° ìœ„í•´ ê²€ì¦ ë°ì´í„°(`X_val`, `y_val`)ì„ ì´ìš©í•˜ì—¬ ROC-AUC ê°’ì„ êµ¬í•´ë³´ì.

```python
from sklearn.metrics import roc_auc_score

y_pred_rf = model_rf.predict_proba(X_val)
y_pred_xgb1 = model_xgb1.predict_proba(X_val)

score_rf = roc_auc_score(y_val, y_pred_rf[:, 1])
score_xgb1 = roc_auc_score(y_val, y_pred_xgb1[:, 1])

print(score_rf)
print(score_xgb1)
```

<div class="alert alert-info">
    ëœë¤ í¬ë ˆìŠ¤íŠ¸ ëª¨ë¸ ê²°ê³¼ëŠ” 0.888, XGBoost ëª¨ë¸ ê²°ê³¼ëŠ” 0.092ë¡œ XGBoost ëª¨ë¸ ì„±ëŠ¥ì´ ì•½ê°„ ë” ì¢‹ê²Œ ë‚˜ì˜¤ëŠ” ê²ƒì„ í™•ì¸í•  ìˆ˜ ìˆë‹¤.

- ì´ ë‘ ëª¨ë¸ì˜ í•˜ì´í¼íŒŒë¼ë¯¸í„° íŠœë‹ì„ í†µí•´ ì„±ëŠ¥ì„ ë” ì˜¬ë ¤ë³´ì.
- ê·¸ ì „ì— ìƒˆë¡œ ë§Œë“  `capital_gain_yn`, `capital_loss_yn`ì´ ëª¨ë¸ ì˜ˆì¸¡ì— ë„ì›€ì„ ì£¼ì—ˆëŠ”ì§€ í™•ì¸í•´ë³´ì.

    capital_gain, capital_loss, capital_gain_yn, capital_loss_yn 4ê°œ ë³€ìˆ˜ì˜ ì¤‘ìš”ë„ë¥¼ ë³´ë©´ ëª¨ë‘ capital_gain, capital_lossê°€ ë” ì¤‘ìš”í•˜ê²Œ ì“°ì„ì„ ì•Œ ìˆ˜ ìˆë‹¤.

- í•˜ì´í¼íŒŒë¼ë¯¸í„° íŠœë‹ ì „ì— `capital_gain_yn`, `capital_loss_yn` ë³€ìˆ˜ë¥¼ ì œì™¸í•˜ë„ë¡ í•œë‹¤.

```python
COL_DEL = ['capital_gain_yn', 'capital_loss_yn']

X_tr = X_tr.drop(COL_DEL, axis=1)
X_val = X_val.drop(COL_DEL, axis=1)
X_test = X_test.drop(COL_DEL, axis=1)
```

- ëœë¤ í¬ë ˆìŠ¤íŠ¸ ëª¨ë¸ ë¨¼ì € í•˜ì´í¼íŒŒë¼ë¯¸í„° íŠœë‹ì„ ì§„í–‰í•´ë³¸ë‹¤.

|í•˜ì´í¼íŒŒë¼ë¯¸í„°|ì„¤ëª…|
|:-|:-|
|`n_estimators`|ê²°ì • íŠ¸ë¦¬ ê°œìˆ˜ (`default=100`)|

- ì´ ì¤‘, `min_sample_split`, `min_sample_leaf`ëŠ” íŠ¹íˆ <u>ê³¼ì í•© ë°©ì§€</u>ë¥¼ ìœ„í•´ ì‚¬ìš©ë˜ëŠ” í•˜ì´í¼íŒŒë¼ë¯¸í„°ì´ë‹¤.
- `GridSearchCV` í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•´ì„œ ëª‡ ê°œì˜ í•˜ì´í¼íŒŒë¼ë¯¸í„° ì¡°í•©ì„ ì‹œë„í•´ë³¼ ìˆ˜ ìˆë‹¤.
