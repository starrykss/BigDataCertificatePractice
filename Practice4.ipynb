## 👨🏻‍💻 문제 풀기

```python
import pandas as pd
import numpy as np
```

```python
X_test = pd.read_csv("./datasets/Part2/census_X_test.csv")
X_train = pd.read_csv("./datasets/Part2/census_X_train.csv")
y_train = pd.read_csv("./datasets/Part2/census_Y_train.csv")
```

### 데이터 전처리

- `info()` 함수를 사용하여 **컬럼 정보, 결측치가 아닌 정상적인 데이터 개수, 데이터 타입**을 확인한다.

```python
print(X_train.info())
```

- 문자열로된 변수는 모델 학습 전 인코딩 작업이 필요하다.
- `describe()` 함수를 통해 **평균, 표준편차, 사분위수** 등 기본적인 **요약 통계량 값**을 확인한다.

```python
print(X_train.describe())
```

<div class="alert alert-info">
    capital_gain, capital_loss 두 변수는 데이터가 오른쪽으로 크게 치우쳐져 있는 것을 확인할 수 있다.
</div>

- 두 변수의 75 백분위수 ~ 100 백분위수 범위를 5 백분위수 단위로 쪼개어서 살펴보자.

```python
print(X_train['capital_gain'].quantile([q / 20 for q in range(15, 21)]))
```

```python
print(X_train['capital_loss'].quantile([q / 20 for q in range(15, 21)]))
```

<div class="alert alert-info">
    capital_gain 변수는 10% 이내, capital_loss 변수는 5% 이내만 0보다 큰 값을 갖는다.
    <br/>
    이 경우, 변수 그대로 사용하거나, 데이터 변환을 하거나, 범주형 변수로 변환하여 사용할 수 있다.
</div>

- 범주형 변수로 변환하여 새로운 파생 변수를 만들고, 초도 모델링 이후 변수 중요도를 통해 기존 변수와 파생 변수 중 어떤 변수가 더 중요하게 쓰이는지 확인해보자.
    - `where` 함수를 이용해서 파생 변수 `capital_gain_yn`, `capital_loss_yn`을 만든다.

```python
X_train['capital_gain_yn'] = np.where(X_train['capital_gain'] > 0, 1, 0)
X_train['capital_loss_yn'] = np.where(X_train['capital_loss'] > 0, 1, 0)

X_test['capital_gain_yn'] = np.where(X_test['capital_gain'] > 0, 1, 0)
X_test['capital_loss_yn'] = np.where(X_test['capital_loss'] > 0, 1, 0)
```

- 종속 변수 값별로 독립 변수 값의 차이가 있는지 탐색하기 위해 변수 종류별로 나눠본다.

```python
COL_DEL = []
COL_NUM = ['age', 'education_num', 'hours_per_week', 'capital_gain', 'capital_loss']    # 수치형 변수
COL_CAT = ['workclass', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'native_country', 'capital_gain_yn', 'capital_loss_yn']    # 범주형 변수
COL_Y = ['target']

X_train = X_train.drop(COL_DEL, axis=1)
X_test = X_test.drop(COL_DEL, axis=1)
```

- 먼저 수치형 변수에 대해 종속 변수 값별로 각 독립 변수의 분포를 확인해보자.

```python
train_df = pd.concat([X_train, y_train], axis=1)

for _col in COL_NUM:
    print("-" * 80)
    print(_col)
    print(train_df.groupby(COL_Y)[_col].describe(), end="\n\n")
```

<div class="alert alert-info">
    age, education_num, hours_per_week 변수 모두 종속 변수 값이 1일 때의 평균이 0일 때의 평균보다 더 크다. <br/>
    capital_gain, capital_loss는 종속 변수 값에 따른 평균 값 차이가 더 크다. <br/>
    수치형 변수 모두 종속 변수를 예측하는 데 사용할 만하다고 보인다.
</div>

- 각 범주형 변수별로 범주값별 종속 변수 평균을 확인해보자.
- 범주값에 따라 종속 변수가 `1`인 비율이 차이가 있는지 간단히 확인하는 방법이다.

```python
for _col in COL_CAT:
    print(train_df.groupby(_col, as_index=False)[COL_Y].mean().sort_values(by=COL_Y, ascending=False), end="\n\n")
```

<div class="alert alert-info">
    학습 데이터에서 종속 변수가 1인 비율은 약 0.249이다. <br/>
    위의 결과를 보면 범주값별로 종속 변수 평균이 다양하게 나타나서 해당 변수는 종속 변수 예측에 사용할 만하다고 볼 수 있다. <br/>
    native country 변수의 경우 41개의 고유한 값을 갖는다. (한 변수가 가질 수 있는 고유한 값의 개수를 <b>카디널리티(Cardinality)</b>라고 한다.) <br/>
    카디널리티가 높은 변수는 모델 성능에 악영향을 준다. (보통 100개 이상의 고유한 값을 가지면 카디널리티가 높다고 한다.) <br/>
    카디널리티가 높을 경우, 해당 변수를 제거하거나 일부 범주를 합쳐 카디널리티를 낮출 수 있다. 또는 (다루기 까다롭지만) 타겟 인코딩을 적용할 수 있다.
</div>

- 범주형 변수들에 대해 **레이블 인코딩**을 수행한다.
    - 레이블 인코딩은 **문자**로 된 범주를 **정수** 형태로 바꿔준다.

```python
from sklearn.preprocessing import LabelEncoder

X = pd.concat([X_train, X_test])
for _col in COL_CAT:
    le = LabelEncoder()
    le.fit(X_train[_col])
    X_train[_col] = le.transform(X_train[_col])
    X_test[_col] = le.transform(X_test[_col])
```

<div class="alert alert-warning">
    레이블 인코딩을 한 데이터는 <u>선형 모델</u>에는 적합하지 않다. (레이블 인코딩 한 변수 값이 의미를 가지는 것이 아니기 때문이다.) <br/>
    따라서 데이터 모형 구축 시, <u>비선형 모델</u>인 <b>랜덤 포레스트</b>와 <b>XGBoost</b> 알고리즘을 사용하도록 한다.
</div>

### 데이터 모형 구축

- 분류 예측 모형을 만들고자 우선 데이터 분할을 진행한다.

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.3, stratify=y_train)
```

- 분할된 데이터로 **수치형 변수**의 데이터 스케일링을 수행한다.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_tr[COL_NUM] = scaler.fit_transform(X_tr[COL_NUM])
X_val[COL_NUM] = scaler.fit_transform(X_val[COL_NUM])
X_test[COL_NUM] = scaler.fit_transform(X_test[COL_NUM])
```

- split이 되었으니 바로 해당 데이터를 사용하여 학습을 진행한다.
- `sklearn` 패키지를 이용하여 랜덤 포레스트, XGBoost 모델을 만들어보자.

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(X_tr, y_tr.values.ravel())
```

```python
from xgboost import XGBClassifier

model_xgb1 = XGBClassifier()
model_xgb1.fit(X_tr, y_tr.values.ravel())
```

<div class="alert alert-warning">
    XGBoost 학습 시, <code>eval_set</code>을 사용하여 충분히 학습을 시키면서 과적합을 방지하도록 할 수 있다.
</div>

```python
model_xgb2 = XGBClassifier(n_estimators=1000, learning_rate=0.1, max_depth=10)
model_xgb2.fit(X_tr, y_tr.values.ravel(), early_stopping_rounds=50, eval_metric='auc', eval_set=[(X_val, y_val)], verbose=10)
```

<div class="alert alert-warning">
    n_estimators 값을 1,000으로 크게 잡을 경우 학습을 더 잘 할 수 있지만, 그만큼 과대적합 위험이 크다. <br/>
    그래서 fitting 할 때 eval_set과 early_stopping_rounds 값을 지정해주었다. <br/>
    모델이 매 iteration 마다 eval_set을 이용하여 성능을 계산하다가 early_stopping_rounds 이상의 iteration 동안 성능 개선이 없으면 학습을 멈춘다.
</div>

- 초도 모델의 성능을 확인하기 위해 검증 데이터(`X_val`, `y_val`)을 이용하여 ROC-AUC 값을 구해보자.

```python
from sklearn.metrics import roc_auc_score

y_pred_rf = model_rf.predict_proba(X_val)
y_pred_xgb1 = model_xgb1.predict_proba(X_val)

score_rf = roc_auc_score(y_val, y_pred_rf[:, 1])
score_xgb1 = roc_auc_score(y_val, y_pred_xgb1[:, 1])

print(score_rf)
print(score_xgb1)
```

<div class="alert alert-info">
    랜덤 포레스트 모델 결과는 0.888, XGBoost 모델 결과는 0.092로 XGBoost 모델 성능이 약간 더 좋게 나오는 것을 확인할 수 있다.
</div>

- 이 두 모델의 하이퍼파라미터 튜닝을 통해 성능을 더 올려보자.
- 그 전에 새로 만든 `capital_gain_yn`, `capital_loss_yn`이 모델 예측에 도움을 주었는지 확인해보자.

```python
pd.DataFrame({'feature': X_tr.columns, 'fi_rf': model_rf.feature_importances_, 'fi_xgb': model_xgb1.feature_importances_})
```

<div class="alert alert-info">
    각 모델의 feature_importances_ 값은 변수 중요도를 나타낸다. <br/>
    capital_gain, capital_loss, capital_gain_yn, capital_loss_yn 4개 변수의 중요도를 보면 모두 capital_gain, capital_loss가 더 중요하게 쓰임을 알 수 있다.
</div>

- 하이퍼파라미터 튜닝 전에 `capital_gain_yn`, `capital_loss_yn` 변수를 제외하도록 한다.

```python
COL_DEL = ['capital_gain_yn', 'capital_loss_yn']

X_tr = X_tr.drop(COL_DEL, axis=1)
X_val = X_val.drop(COL_DEL, axis=1)
X_test = X_test.drop(COL_DEL, axis=1)
```

- 랜덤 포레스트 모델 먼저 하이퍼파라미터 튜닝을 진행해본다. 
- 랜덤 포레스트의 주요 하이퍼파라미터는 다음과 같다.

|하이퍼파라미터|설명|
|:-|:-|
|`n_estimators`|결정 트리 개수 (`default=100`)|
|`max_depth`|트리 깊이의 최댓값 (`default=None`)|
|`min_sample_split`|노드 분리에 필요한 최소 샘플 수 (`default=2`)|
|`min_sample_leaf`|리프 노드에 필요한 최소 샘플 수 (`default=1`)|
|`max_features`|최적의 분리를 위해 고려할 변수 개수 (`default=auto`)|

- 이 중, `min_sample_split`, `min_sample_leaf`는 특히 <u>과적합 방지</u>를 위해 사용되는 하이퍼파라미터이다.
- `GridSearchCV` 함수를 사용해서 몇 개의 하이퍼파라미터 조합을 시도해볼 수 있다.
    - `estimator` : 튜닝에 사용할 기본 모델
    - `params_grid` : 하이퍼파라미터 조합
    - `cv` : 교차 검증 시 몇 개의 fold를 사용할지

```python
from sklearn.model_selection import GridSearchCV

grid_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf_cv = GridSearchCV(estimator=model_rf, param_grid=grid_params, cv=5)
rf_cv.fit(X_train, y_train.values.ravel())
```

<div class="alert alert-info">
    grid_params라는 딕셔너리에 여러 개의 하이퍼파라미터 조합을 저장했다. 이 경우, 3x3x3=81개 경우의 수를 테스트한다. <br/>
    params_grid에 너무 많은 조합을 넣으면 튜닝 시간이 너무 오래 걸릴 수 있으므로 주의한다.
</div>
<div class="alert alert-danger">
    실제로 시험을 응시할 때 전체 코드 실행 시간은 <u>1분 제한</u> 조건이 존재하기 때문에, 이 코드는 정확도를 높이는 데엔 효과적일 수 있지만, 시험에 제출을 할 경우 제대로 제출이 되지 않을 확률이 매우 높다. <br/>
    수행 시간이 너무 오래 걸리는 문제가 발생한다면, 하이퍼파라미터를 2~3개씩 나눠서 튜닝을 수행할 수도 있다.
</div>

- `cv_results_` 속성을 불러오면 각각 사용한 파라미터 값과 그 때의 fold 별 성능과 평균 성능 등을 모두 확인할 수 있다.

```python
print(pd.DataFrame(rf_cv.cv_results_).head())
```

- `best_estimator_` 속성을 사용하면 그리드 서치 중 가장 좋은 성능을 보였던 estimator를 불러올 수 있다.
- 하지만, 여기서는 `best_params_` 속성을 불러와 가장 좋은 성능을 보인 하이퍼파라미터 조합으로 새로 모델을 생성하였다.

```python
print(rf_cv.best_params_)
```

- 이번에는 XGBoost 모델의 하이퍼파라미터 튜닝을 진행해본다.
- XGBoost 모델의 주요 하이퍼파라미터는 다음과 같다.

|하이퍼파라미터|설명|
|:-|:-|
|`n_estimators`|부스팅 iteration 수. 일반적으로 100~1000 사이의 값을 가짐. (`default=100`)|
|`max_depth`|트리 깊이의 최댓값. 일반적으로 3~10 사이의 값을 가짐. (`default=3`)|
|`learning_rate`|학습률. 일반적으로 0.01~0.2 사이의 값을 가짐. (`default=0.1`)|
|`min_child_weigt`|자식 노드에 필요한 가중치 합의 최솟값. `min_child_leaf`와 유사 (`default=1`)|
|`subsample`|각 트리에 사용할 데이터 샘플 비율. 일반적으로 0.5~1 사이의 값을 가짐. (`default=1`)|
|`colsample_bytree`|각 트리에 사용할 변수 비율. 일반적으로 0.5~1 사이의 값을 가짐. (`default=1`)|
|`lambda`|L2 규제 값 (`default=1`)|
|`alpha`|L1 규제 값 (`default=0`)|

- 이 중, `max_depth` 값을 가장 주의 깊게 봐야 한다.
    - `max_depth` 값이 너무 작으면 과소적합, 너무 크면 과대적합의 위험이 있다.
- `min_child_weight` 값을 통해 과대적합을 방지할 수 있다.

```python
grid_params = {'max_depth': [3, 5, 7, 10],
               'min_child_weight': [1, 2],
               'colsample_bytree': [0.6, 0.8],
               'subsample': [0.6, 0.8]}

xgb_cv = GridSearchCV(estimator=model_xgb1, param_grid=grid_params, cv=5)
xgb_cv.fit(X_tr, y_tr.values.ravel())
```

- `best_params_` 속성을 불러와 가장 좋은 성능을 보인 하이퍼파라미터 조합으로 새로 모델을 생성한다.

```python
print(xgb_cv.best_params_)
```

```python
params = {'colsample_bytree': 0.6,
          'max_depth': 7,
          'min_child_weight': 1,
          'subsample': 0.8}

model_xgb3 = XGBClassifier(n_estimators=1000, learning_rate=0.05)
model_xgb3.set_params(**params)

model_xgb3.fit(X_tr, y_tr, early_stopping_rounds=50, eval_metric='auc', eval_set=[(X_val, y_val)], verbose=10)

print(model_xgb3.best_score)
```
